# Finding outliers

Z-Score

```python
from scipy import stats
```

```python
df1 = df.drop(columns=['Country', 'Currency'])
z_scores = np.abs(stats.zscore(df1))
z_scores
```

```python
# if z-score > 3 or < -3, it will be considered an outlier
df_no_outlier = df[(z_scores < 3).all(axis=1)]
df_no_outlier.head()
```

```python
print(f"Rows and columns in df_no_outlier: {df_no_outlier.shape}\nRows and Columns in the original df: {df.shape}")
```

IQR (Interquartile Range)

```python
q1 = df.quantile(.25)
q3 = df.quantile(.75)
iqr = q3 - q1
```

```python
df_iqr_no_outliers = df[~((df < (q1 - 1.5*iqr)) | (df > (q3 + 1.5*iqr)))]
```

```python
print(f"Number of rows and columns in df_iqr_no_outliers: {df_iqr_no_outliers.shape}\nNumber of rows and columns in original df: {df.shape}")
```
